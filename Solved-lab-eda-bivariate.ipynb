```python
import pandas as pd
import numpy as np
from scipy.stats import chi2_contingency
import matplotlib.pyplot as plt

data = pd.DataFrame({
    'category': ['Hi-Fi Speakers', 'Hi-Fi Speakers', 'Hi-Fi Speakers', 'Hi-Fi Speakers', 'Hi-Fi Speakers',
                 'Motorbike Clothing', 'Motorbike Clothing', 'Motorbike Clothing', 'Motorbike Clothing'],
    'isBestSeller': [False, True, False, False, False, False, False, False, False]
})

# Crosstab entre categorías y estado de "Best-Seller"
crosstab = pd.crosstab(data['category'], data['isBestSeller'])
crosstab_proportions = crosstab.div(crosstab.sum(axis=1), axis=0).sort_values(by=True, ascending=False)

# Chi-square test
chi2, p, dof, expected = chi2_contingency(crosstab)

# Cramér's V computation
n = crosstab.sum().sum()
cramers_v = np.sqrt(chi2 / (n * (min(crosstab.shape) - 1)))

# Mostrar resultados
results = {
    "Chi-square Test Statistic": chi2,
    "P-value": p,
    "Degrees of Freedom": dof,
    "Cramér's V": cramers_v
}
crosstab_proportions
```

```python
# Crear gráfico de barras apiladas
crosstab.plot(kind='bar', stacked=True, figsize=(8, 6), color=['lightblue', 'orange'])

# Añadir etiquetas y título
plt.title('Relationship Between Product Categories and Best-Seller Status', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Count', fontsize=12)
plt.legend(title='Is Best-Seller', labels=['No', 'Yes'], fontsize=10)
plt.xticks(rotation=45)
plt.tight_layout()

# Mostrar el gráfico
plt.show()
```

```python
# Generating a synthetic dataset similar to the provided image for analysis
data = pd.DataFrame({
    'category': ['Hi-Fi Speakers', 'Hi-Fi Speakers', 'Hi-Fi Speakers', 'Motorbike Clothing', 'Motorbike Clothing'] * 200,
    'price': [21.99, 23.99, 31.99, 22.34, 97.99] * 200,
    'stars': [4.7, 4.7, 4.7, 0.0, 4.3] * 200
})

# Step 1: Remove outliers in the 'price' column using the IQR method
Q1 = data['price'].quantile(0.25)
Q3 = data['price'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter out outliers
data_no_outliers = data[(data['price'] >= lower_bound) & (data['price'] <= upper_bound)]

# Step 2: Violin Plot for price across categories (top 20 categories based on count)
import seaborn as sns

# Top 20 categories based on count
top_categories = data_no_outliers['category'].value_counts().head(20).index
filtered_data = data_no_outliers[data_no_outliers['category'].isin(top_categories)]

# Violin plot
plt.figure(figsize=(12, 6))
sns.violinplot(x='category', y='price', data=filtered_data, palette="muted", scale="width")
plt.title('Distribution of Prices Across Top 20 Categories', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Price', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()
plt.show()

# Step 3: Bar Chart for average price in top 10 categories by count
top_10_categories = data_no_outliers['category'].value_counts().head(10).index
average_prices = data_no_outliers[data_no_outliers['category'].isin(top_10_categories)].groupby('category')['price'].mean()

# Bar chart
plt.figure(figsize=(10, 6))
average_prices.sort_values(ascending=False).plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Average Price of Products in Top 10 Categories', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Average Price', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()
plt.show()

# Step 4: Box Plot for ratings across top 10 categories by count
plt.figure(figsize=(12, 6))
sns.boxplot(x='category', y='stars', data=filtered_data, palette="muted")
plt.title('Distribution of Ratings Across Top 10 Categories', fontsize=14)
plt.xlabel('Category', fontsize=12)
plt.ylabel('Rating', fontsize=12)
plt.xticks(rotation=45, fontsize=10)
plt.tight_layout()
plt.show()

# Results on median/highest price and ratings
median_price_per_category = data_no_outliers.groupby('category')['price'].median().sort_values(ascending=False)
highest_rating_per_category = data_no_outliers.groupby('category')['stars'].median().sort_values(ascending=False)

median_price_per_category.head(1), highest_rating_per_category.head(1)
```

```python
import scipy.stats as stats

# Step 1: Calculate the correlation coefficient between price and stars
correlation_coefficient = data_no_outliers[['price', 'stars']].corr().iloc[0, 1]

# Step 2: Scatter plot of product rating vs. price
plt.figure(figsize=(8, 6))
sns.scatterplot(x='stars', y='price', data=data_no_outliers, alpha=0.7, edgecolor=None)
plt.title('Scatter Plot: Product Rating vs. Price', fontsize=14)
plt.xlabel('Product Rating (Stars)', fontsize=12)
plt.ylabel('Product Price', fontsize=12)
plt.tight_layout()
plt.show()

# Step 3: Correlation heatmap
plt.figure(figsize=(6, 4))
sns.heatmap(data_no_outliers.corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap for Numerical Variables', fontsize=14)
plt.tight_layout()
plt.show()

# Step 4: Check if product prices follow a normal distribution using a QQ plot
plt.figure(figsize=(6, 6))
stats.probplot(data_no_outliers['price'], dist="norm", plot=plt)
plt.title('QQ Plot for Product Prices', fontsize=14)
plt.tight_layout()
plt.show()

# Bonus: Correlation analysis with outliers
correlation_with_outliers = data[['price', 'stars']].corr().iloc[0, 1]

(correlation_coefficient, correlation_with_outliers)
```
